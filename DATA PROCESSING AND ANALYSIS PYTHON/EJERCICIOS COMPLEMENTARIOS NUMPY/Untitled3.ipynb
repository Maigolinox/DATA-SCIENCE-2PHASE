```python
import math
nums=range(361)
for i in nums:
    print("Análisis de posición con "+str(i)+" Grados")
    AD=10*math.sin(i)
    print("El AD es: "+str(AD))
    BD=10*math.sin(i)
    print("El BD es: "+str(BD))
    DC=math.sqrt((25**2)-(BD**2))
    print("El DC es: "+str(DC))
    AC=AD+DC
    print("El AC es: "+str(AC))
    DESPL=AC-15
    print("El desplazamiento es: "+str(DESPL))
```
